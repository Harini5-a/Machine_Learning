```python
import pandas as pd
import numpy as np
import math
import random

from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv('/content/drive/MyDrive/data.csv')
data = data.drop(columns=["RollNo"])
data = data.sample(frac=1).reset_index(drop=True)

split = int(0.8 * len(data))
traindata = data[:split]
testdata = data[split:]
target = data.columns[-1]

features = list(data.columns)
features.remove(target)

def classcounts(data, target):
    counts = {}
    for index, row in data.iterrows():
        label = row[target]
        if label not in counts:
            counts[label] = 1
        else:
            counts[label] += 1
    return counts

def priorprobability(data, target):
    counts = classcounts(data, target)
    total = len(data)
    prior = {}
    for label in counts:
        prior[label] = counts[label] / total
    return prior

def likelihood(data, feature, value, label, target):
    classdata = data[data[target] == label]
    numerator = len(classdata[classdata[feature] == value])
    denominator = len(classdata)
    return (numerator + 1) / (denominator + data[feature].nunique())

def predict(traindata, sample, target):
    prior = priorprobability(traindata, target)
    probability = {}
    for label in prior:
        result = prior[label]
        for feature in features:
            value = sample[feature]
            result *= likelihood(traindata,feature,value,label,target)
        probability[label] = result
    prediction = max(probability, key=probability.get)
    return prediction
predictions = []

for index, row in testdata.iterrows():
    prediction = predict(traindata, row, target)
    predictions.append(prediction)
testdata["Predicted"] = predictions
print(testdata)
```
